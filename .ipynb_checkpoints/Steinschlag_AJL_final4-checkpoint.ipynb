```python
import numpy as np
import pandas as pd
import scipy.stats as st
import scipy
import matplotlib
import matplotlib.pyplot as plt
import os.path

matplotlib.style.use('ggplot')
```

```python
# the datafiles and the notebook need to be inside the same folder
# usecols, to discard the empty columns
# skipfooter, to discard rows containing nothing
path1 = os.path.abspath('out_1.csv')
path2 = os.path.abspath('out_2.csv')
zone1 = pd.read_csv(path1, usecols=[0, 1, 2, 3], skipfooter=11, parse_dates={'Date_Time':[0,1]}, engine='python')
zone2 = pd.read_csv(path2, usecols=[0, 1, 2, 3], skipfooter=3, parse_dates={'Date_Time':[0,1]}, engine='python')

# renaming the columns for clarity and uniformity
newNames1 = {'Masse [kg]' : 'm [kg]', 'Geschwindigkeit [m/s]' : 'v [m/s]'}
zone1.rename(columns=newNames1, inplace=True)

# sorting, so that the timedelta variables will be correct
zone1.sort_values('Date_Time', inplace = True, ignore_index = True)
zone2.sort_values('Date_Time', inplace = True, ignore_index = True)
```

```python
# creating timedelta-variables that hold the time differences betwen two stones
timedelta1 = []
index1 = 1
index2 = 0    
while index1 <= 67:
    timedelta1.append((zone1['Date_Time'].iloc[index1] - zone1['Date_Time'].iloc[index2]) / pd.Timedelta(seconds = 1))
    index1 += 1
    index2 += 1
    
    
timedelta2 = []
index1 = 1
index2 = 0    
while index1 <= 31:
    timedelta2.append((zone2['Date_Time'].iloc[index1] - zone2['Date_Time'].iloc[index2]) / pd.Timedelta(seconds = 1))
    index1 += 1
    index2 += 1
```

```python
# returns a list with distribution name, parameters, pdf and cdf
def get_best_fit_distribution(data, size_for_cdf, preferred_distributions = True):
    
    y, x = np.histogram(data, density=True)
    x = (x + np.roll(x, -1))[:-1] / 2.0
    x2 = np.arange(size_for_cdf) # for a better graph a bigger nummber of x-values is needed
        
    preferred_distribution_names = ['expon', 'gamma', 'lognorm', 'norm']
    possible_distribution_names = ['beta', 'cauchy', 'chi2', 'expon', 'f', 'gamma', 'logistic', 'lognorm', 'norm', 'pareto', 't', 'uniform']
    
    # holders for best values
    best_distribution = None
    best_parameters = None
    best_sse = np.inf
    
    if preferred_distributions:
        for distribution_name in preferred_distribution_names:
            distribution = getattr(scipy.stats, distribution_name)
            parameters = distribution.fit(data)
            pdf = distribution.pdf(x, *parameters[:-2], loc=parameters[-2], scale=parameters[-1])
            sse = np.sum(np.power(y - pdf, 2.0))
            
            
            # distribution test
            if best_sse > sse > 0:
                best_distribution = distribution_name
                best_parameters = parameters
                best_sse = sse
                
    else:
        for distribution_name in possible_distribution_names:
            distribution = getattr(scipy.stats, distribution_name)
            parameters = distribution.fit(data)
            pdf = distribution.pdf(x, *parameters[:-2], loc=parameters[-2], scale=parameters[-1])
            sse = np.sum(np.power(y - pdf, 2.0))
            
            # distribution test
            if best_sse > sse > 0:
                best_distribution = distribution_name
                best_parameters = parameters
                best_sse = sse
    
    best_pdf = getattr(scipy.stats, best_distribution).pdf(x2, *best_parameters[:-2], loc = best_parameters[-2], scale = best_parameters[-1])
    best_cdf = getattr(scipy.stats, best_distribution).cdf(x2, *best_parameters[:-2], loc = best_parameters[-2], scale = best_parameters[-1])
    
    results = [best_distribution, best_parameters, best_pdf, best_cdf]
    
    return(results)
```

```python
# return empirical distribution function
def get_edf(data):
    edf_x = np.sort(data)
    edf_y = np.arange(1, len(data) + 1) / len(data)
    return(edf_x, edf_y)
```

```python
def plot_results(edf_x, edf_y, results, xlabel, title):
    plt.plot(results[-1], label = results[0] + " cdf")
    plt.plot(edf_x, edf_y , label = 'edf')
    plt.xlabel(xlabel)
    plt.ylabel("P(X<x)")
    plt.title(title)
    plt.legend(loc = 'lower right')
    plt.show()
```

```python
mass1 = zone1['m [kg]']
results_mass1 = get_best_fit_distribution(mass1, 4000)
edf_x, edf_y = get_edf(mass1)   
plot_results(edf_x, edf_y, results_mass1, "x = Masse [kg]", "Beste Verteilung f√ºr Masse Zone 1")
```

```python
def death_probability_speed(crash_speed):
    mean_carspeed = 70
    variance_carspeed = 20

    death_probability = st.norm.cdf(crash_speed, mean_carspeed, variance_carspeed)
    
    return death_probability

#plot of curve of normal distribution cumulative distribution function
mean_carspeed = 70
variance_carspeed = 20
x = np.linspace(0, 120, 100)
y_cdf = st.norm.cdf(x, mean_carspeed, variance_carspeed) # the normal cdf
plt.xlabel("Aufprallgeschwindigkeit(km/h)")
plt.ylabel("Todeswahrsceinlichkeit")
plt.title("Todeswahrschinlichkeit in Abh√§ngigkeit der Aufprallgeschwindigkeit")
plt.plot(x, y_cdf, label='cdf')
plt.show()
#y-achse einer CDF: P(X<x) wobei x = wert auf x achse
```
